```python
import os
import configparser

from langchain_groq import ChatGroq
from langchain_cohere import ChatCohere

from langchain_core.messages import HumanMessage, SystemMessage

config = configparser.ConfigParser()
config.read('/workspaces/langchainCDAC9nov/config.ini')
groq = config['groq']
cohere = config['cohere']

os.environ['GROQ_API_KEY'] = groq.get('GROQ_API_KEY')
os.environ['COHERE_API_KEY'] = cohere.get('COHERE_API_KEY')

messages = [
    SystemMessage(content='You are a weather service. You will respond to weather queries to the best of you ability. You will always end with - Have a great day'),
    HumanMessage(content='Hey whats the weather like today?')
]



## code for cohere.
model = ChatCohere(model="command-r-plus")
print(model.invoke(messages))

## Code for Groq
model = ChatGroq(model="llama3-8b-8192")
print(model.invoke(messages))



```

```python
import random

def actor_picker():
  actors = ["Brad Pitt",
"Shahid Kapoor",
"Chris Evans",
"Ranveer Singh",
"Benedict Cumberbatch",
"Hrithik Roshan",
"Tom Hardy",
"Vicky Kaushal"]
  return random.choice(actors)

print("Actor:",actor_picker())

def location_picker():
  location=["India","Italy","Dubai","Japan","Canada","New Zealand","Paris","France","Greece"]
  return random.choice(location)

print("Location:",location_picker())

def theme_picker():
  theme=["Comedy","Horror","Action","Drama","Adventure","Thriller","Science Friction","Fantasy","Animation","Mystery"]
  return random.choice(theme)

print("Theme:",theme_picker())

# def story():
#   story=[actor_picker(),location_picker(),theme_picker()]
#   return story

# print("Story:",story())
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate #import the Class for creating a prompt
from langchain_groq import ChatGroq
parser = StrOutputParser()

Actor=actor_picker()
Location=location_picker()
Theme=theme_picker()

# templatized system prompt
system_template = "On the basis of actor,location and theme create a random movie story. Please provide a {responseType} response."

# Create prompt template instance.
prompt_template = ChatPromptTemplate.from_messages(
    [
        ("system", system_template),
        ("user", Actor),
        ("user", Location),
        ("user", Theme)
    ]
)


# Groq
# print("chat from groq")
model = ChatGroq(model="llama3-8b-8192")
chain = prompt_template | model | parser
chain.invoke({"responseType":"descriptive"})
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate #import the Class for creating a prompt
from langchain_cohere import ChatCohere


parser = StrOutputParser()

Actor=actor_picker()
Location=location_picker()
Theme=theme_picker()

# templatized system prompt
system_template = "On the basis of actor,location and theme create a random movie story. Please provide a {responseType} response."

# Create prompt template instance.
prompt_template = ChatPromptTemplate.from_messages(
    [
        ("system", system_template),
        ("user", Actor),
        ("user", Location),
        ("user", Theme)
    ]
)

model = ChatCohere(model="command-r-plus")
coherechain = prompt_template | model | parser
cohere=coherechain.invoke({"responseType":"descriptive"})
print(cohere)


```

```python
from langchain_core.runnables.history import RunnableWithMessageHistory
store = {}

def get_session_history(session_id: str) -> BaseChatMessageHistory:
    if session_id not in store:  # If a new session then create a new memory store.
        store[session_id] = InMemoryChatMessageHistory()
    return store[session_id]
config = {'configurable': {"session_id": "123abc"}}
withHistory = RunnableWithMessageHistory(model, get_session_history)

response = withHistory.invoke([HumanMessage(content=cohere)], config=config)

print(response.content) # all good so far

# we dont need to explicitly store the response from the model in history

response = withHistory.invoke(
    [HumanMessage(content="name of the actor ")], config=config
)

print(response.content)
```
